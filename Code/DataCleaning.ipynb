# Dataset Cleaning
The MDP dataset is in .arff format. These will be converted to .csv format for easier manipulation.

```python
sample_arff_data = arff.loadarff("../Datasets/MDP/CM1.arff")
sample_csv_data = pd.DataFrame(sample_arff_data[0])
sample_csv_data
```

```python
import os
```

```python
arff_filenames = os.listdir("../Datasets/MDP/")
arff_filenames
```

```python
for file_name in arff_filenames:
    data = arff.loadarff("../Datasets/MDP/" + file_name)
    df = pd.DataFrame(data[0])
    df.to_csv("../Datasets/MDP_CSV/" + file_name[:-5] + ".csv", index=False)
```
